```python
# Import libraries
import json, csv
import pandas as pd
```

```python
# Open JSON file store to data_file
data_file = open('yelp_academic_dataset_review.json', encoding='utf-8')

# Initialize an empty list to store data
data = []

# Read each line in the file and append as JSON object to the list
for line in data_file:
    data.append(json.loads(line))
    
# Create a dataframe using the collected data
review_df = pd.DataFrame(data)
data_file.close()

# Open JSON file store to data_file
data_file = open('yelp_academic_dataset_business.json', encoding='utf-8')

# Initialize an empty list to store data
data = []

# Read each line in the file and append as JSON object to the list
for line in data_file:
    data.append(json.loads(line))

# Create a dataframe using the collected data
business_df = pd.DataFrame(data)
data_file.close()
```

```python
# List of DataFrame
dfs = [review_df, business_df] 
# Loop through each DataFrame and display first 5 rows for each Dataframe
for df in dfs:
    display(df.head(5))
    display(df.info())
```

```python
# Merge dataframe on 'business_id' column using outer join
merged_df = pd.merge(business_df, review_df, on ='business_id', how = 'outer')
merged_df.head()
```

```python
# Display merged_df summary
merged_df.info()
```

```python
# Rename the columns
merged_df.rename({"stars_x": "business_stars", "stars_y": "user_stars"}, axis = "columns", inplace = True)

# Filter merged_df with 'name' column keep row with 'Starbuck' only
df_starbucks = merged_df.loc[merged_df['name'] == 'Starbucks']\

# Display summary of df_starbuck
df_starbucks.head()
```

```python
# Save df_starbuck in to csv file.
df_starbucks.to_csv('starbucks_df', index = False)
```
